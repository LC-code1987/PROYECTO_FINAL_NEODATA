```python
# Paso 1: Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 2: Cargar el dataset (ajusta la ruta si es necesario)
df = pd.read_csv("ESTADISTICAS_EN_EDUCACION.csv")

# Paso 3: Estandarizar nombres de columnas
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Paso 4: Mostrar información general
print("Dimensiones:", df.shape)
df.info()

# Paso 5: Verificar porcentaje de nulos
porcentaje_nulos = df.isnull().mean() * 100
print("\nPorcentaje de valores nulos por columna:\n", porcentaje_nulos)

# Paso 6: Visualizar columnas con nulos
porcentaje_nulos_filtrado = porcentaje_nulos[porcentaje_nulos > 0]
plt.figure(figsize=(12, 6))
sns.barplot(x=porcentaje_nulos_filtrado.values, y=porcentaje_nulos_filtrado.index, palette="viridis")
plt.title("Porcentaje de valores nulos por columna")
plt.xlabel("Porcentaje")
plt.ylabel("Columnas")
plt.show()

# Paso 7: Eliminar columnas con más de 40% de nulos
columnas_a_eliminar = porcentaje_nulos[porcentaje_nulos > 40].index.tolist()
df.drop(columns=columnas_a_eliminar, inplace=True)
print("Columnas eliminadas por alto porcentaje de nulos:", columnas_a_eliminar)

# Paso 8: Imputar columnas con nulos (menos de 40%)
columnas_a_imputar = porcentaje_nulos[(porcentaje_nulos > 0) & (porcentaje_nulos <= 40)].index.tolist()
for col in columnas_a_imputar:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(df[col].median())  # numéricas
    else:
        df[col] = df[col].fillna(df[col].mode()[0])  # categóricas
print("Columnas imputadas:", columnas_a_imputar)

# Paso 9: Verificación final
print("\nValores nulos restantes:\n", df.isnull().sum().sum())
```

```python
# Paso 1: Calcular porcentaje de nulos
porcentaje_nulos = df.isnull().mean() * 100

# Paso 2: Visualizar (opcional)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
sns.barplot(x=porcentaje_nulos[porcentaje_nulos > 0].values,
            y=porcentaje_nulos[porcentaje_nulos > 0].index,
            hue=porcentaje_nulos[porcentaje_nulos > 0].index,
            dodge=False,
            palette="crest")
plt.title("📉 Porcentaje de valores nulos por columna")
plt.xlabel("Porcentaje")
plt.ylabel("Columnas")
plt.legend([],[], frameon=False)  # Oculta leyenda innecesaria
plt.tight_layout()
plt.show()

# Paso 3: Eliminar columnas con más del 40% de nulos
columnas_muy_nulas = porcentaje_nulos[porcentaje_nulos > 40].index.tolist()
df.drop(columns=columnas_muy_nulas, inplace=True)
print("🗑️ Columnas eliminadas:", columnas_muy_nulas)

# Paso 4: Imputar columnas con menos del 40% de nulos
columnas_imputar = porcentaje_nulos[(porcentaje_nulos > 0) & (porcentaje_nulos <= 40)].index.tolist()
for col in columnas_imputar:
    if df[col].dtype == 'float64' or df[col].dtype == 'int64':
        df[col] = df[col].fillna(df[col].median())  # Numéricas
    else:
        df[col] = df[col].fillna(df[col].mode()[0])  # Categóricas
print("🧩 Columnas imputadas:", columnas_imputar)

# Paso 5: Verificación final
print("\n✅ Nulos restantes en el dataset:", df.isnull().sum().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo gráfico
sns.set(style="whitegrid")

# 🔹 1. Distribución de registros por año
plt.figure(figsize=(10, 4))
sns.countplot(x='año', data=df, palette='Blues_d')
plt.title("Distribución de registros por año")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.tight_layout()
plt.show()

# 🔹 2. Municipios con más registros
top_municipios = df['municipio'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_municipios.values, y=top_municipios.index, palette='crest')
plt.title("🏙️ Top 10 Municipios con más registros")
plt.xlabel("Cantidad de registros")
plt.ylabel("Municipio")
plt.tight_layout()
plt.show()

# 🔹 3. Distribución de tasa de matrícula (5-16 años) por año
# Agrupar por año y calcular la mediana
mediana_matricula = df.groupby('año')['tasa_matriculación_5_16'].median()

plt.figure(figsize=(10, 5))
sns.lineplot(x=mediana_matricula.index, y=mediana_matricula.values, marker='o', color='darkblue')
plt.title("📉 Mediana de tasa de matrícula (5-16 años) por año")
plt.xlabel("Año")
plt.ylabel("Mediana de tasa de matrícula")
plt.grid(True)
plt.tight_layout()
plt.show()
```
